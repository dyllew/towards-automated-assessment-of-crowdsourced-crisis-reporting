```python
import pandas as pd
from os.path import join
from url_image_module import (
    keep_relevant_df_columns,
    drop_empty_df_rows,
    get_label_cols_for_task,
    save_df,
    insert_part_of_path
)
```

```python
DATA_DIR_F = "/content/drive/MyDrive/05_REACT AI/Research/AI & ML/Datasets/Fukuchiyama City Images"
LABELED_IMAGE_DATA_CSV_PATH = join(DATA_DIR_F, 'image_labels_fukuchiyama.csv')
```

```python
task_names = ['damage_severity', 'humanitarian_categories', 'flood_presence', 'informativeness']
relevant_columns_prefixes = ['file_path']
relevant_columns_prefixes.extend(task_names)
```

```python
labeled_data_df = pd.read_csv(LABELED_IMAGE_DATA_CSV_PATH, header=1)
labeled_data_df = keep_relevant_df_columns(labeled_data_df, relevant_columns_prefixes)
labeled_data_df = drop_empty_df_rows(labeled_data_df, how = 'all')
labeled_data_df['file_path'] = labeled_data_df['file_path'].astype(str)
```

```python
labeled_data_df['file_path'] = labeled_data_df['file_path'].apply(lambda x: insert_part_of_path(x, 'damage_photos', 1))
```

```python
ordered_columns = ['file_path'] + [col for task_name in task_names for col in get_label_cols_for_task(labeled_data_df, task_name)]
```

```python
labeled_data_df = labeled_data_df[ordered_columns]
```

```python
# Drop rows which have any NaN values, i.e. those which every category has not been labeled by labelers
labeled_data_df = drop_empty_df_rows(labeled_data_df, how='any')
```

```python
save_df(labeled_data_df, join(DATA_DIR_F, 'FC_labels_snapshot.csv'), encoding='utf-8', index=False)
```
